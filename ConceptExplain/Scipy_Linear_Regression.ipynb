```python
#Feature Function: polyfit and stats.linregress
from scipy import linspace, polyval, polyfit, sqrt, stats, randn
from pylab import plot, title, show , legend

n = 10
x = linspace(0,1,n) #evenly have 50 numbers between -10 and 10

slope = -3; 
intercept = -5
y = polyval([slope,intercept], x) # slope*x**(1) + intercept*x**(0) => slope*x + intercept

yn = y + randn(10) #add some noise randn(n)

print "original data:"
print "n = "+str(n)
print "slope = "+str(slope)
print "intercept = "+str(intercept)
print "x-axis:"+str(x)
print "y-axis:"+str(y)
print "\n"

#Linear regressison -polyfit - polyfit can be used other orders polys
(slope_r,intercept_r)=polyfit(x,yn,1)
y_polyfit_r = polyval([slope_r,intercept_r],x)
#compute the mean square error
err=sqrt(sum((y_polyfit_r - yn)**2)/n)
print('polyfit:\n slope=%.2f intercept=%.2f, ms error= %.3f' % (slope_r,intercept_r,err))
# matplotlib ploting
title('Linear Regression - Polyfit')
plot(x,y,'g.--')
plot(x,yn,'k.')
plot(x,y_polyfit_r,'r.-')
legend(['original','plus noise', 'regression'])
show()
print "y_polyfit_r:"+str(y_polyfit_r)
print "\n"

print('stats.linregress:')
#Linear regression using stats.linregress
(slope2,intercept2,r,tt,stderr)=stats.linregress(x,yn)
print('slope=%.2f intercept=%.2f, std error= %.10f' % (slope2,intercept2,stderr))
y2 = slope2 * x + intercept2
# matplotlib ploting again
title('Linear Regression - stats.linregress')
plot(x,y,'g.--')
plot(x,yn,'k.')
plot(x,y2,'r.-')
legend(['original','plus noise', 'regression'])
show()
print "y2:" + str(y2)
print "\n"

print('polyfit:')
(slope3, intercept3)=polyfit(x,y2,1)
y3 = polyval([slope3,intercept3],x)
err3=sqrt(sum((y3 - y2)**2)/n)
print('slope=%.2f intercept=%.2f, ms error= %.3f' % (slope3,intercept3,err3))
print "y3:"+str(y3)
print "\n"
```
